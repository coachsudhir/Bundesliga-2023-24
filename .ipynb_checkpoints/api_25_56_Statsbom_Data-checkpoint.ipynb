```python
# =========================
# 1. IMPORTS
# =========================
from statsbombpy import sb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx

pd.set_option("display.max_columns", None)


# =========================
# 2. LOAD MATCHES
# =========================
matches = sb.matches(competition_id=9, season_id=281)

germany_matches = matches[
    (matches["home_team"] == "Germany") |
    (matches["away_team"] == "Germany")
]

print("Matches found:", len(germany_matches))


# =========================
# 3. LOAD EVENTS (ALL MATCHES)
# =========================
all_events = []

for match_id in germany_matches["match_id"]:
    ev = sb.events(match_id=match_id)
    ev["match_id"] = match_id
    all_events.append(ev)

events = pd.concat(all_events, ignore_index=True)


# =========================
# 4. FILTER GERMANY EVENTS
# =========================
germany = events[events["team"] == "Germany"].copy()
germany = germany.sort_values(["match_id", "period", "timestamp"])

germany["possession_id"] = (
    germany["match_id"].astype(str) + "_" +
    germany["possession"].astype(str)
)


# =========================
# 5. DEFINE BUILD-UP EVENTS
# =========================
build_actions = ["Pass", "Carry", "Ball Receipt*", "Dribble"]

build_up = germany[germany["type"].isin(build_actions)].copy()


# =========================
# 6. BUILD-UP SEQUENCE FILTER
# =========================
build_up = build_up.groupby("possession_id").filter(lambda x: len(x) >= 3)


# =========================
# 7. KPI METRICS
# =========================
total_sequences = build_up["possession_id"].nunique()

avg_sequence_length = build_up.groupby("possession_id").size().mean()

player_involvement = build_up["player"].value_counts().head(10)

event_share = build_up["type"].value_counts(normalize=True) * 100


# =========================
# 8. PRINT REPORT (TEXT OUTPUT)
# =========================
print("\n==============================")
print("🇩🇪 GERMANY BUILD-UP REPORT")
print("==============================\n")

print(f"Total Build-up Sequences: {total_sequences}")
print(f"Average Sequence Length: {avg_sequence_length:.2f}\n")

print("Event Distribution (%):")
print(event_share)

print("\nTop Players in Build-up:")
print(player_involvement)


# =========================
# 9. SPATIAL HEATMAP
# =========================
if "location" in build_up.columns:
    build_up = build_up.dropna(subset=["location"])
    
    build_up["x"] = build_up["location"].apply(lambda l: l[0])
    build_up["y"] = build_up["location"].apply(lambda l: l[1])

    plt.figure(figsize=(10, 6))
```
